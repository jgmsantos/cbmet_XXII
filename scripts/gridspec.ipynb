# Criação de painel de figuras 

O ```gridspec``` é útil quando se deseja criar um painel de figuras que pode ser personalizada pelo usuário.

[Documentação do GridSpec](https://matplotlib.org/stable/api/_as_gen/matplotlib.gridspec.GridSpec.html)

# Importação de bibliotecas

````python
import matplotlib.pyplot as plt
````

# Entendendo o gridspec

````python
# Define o tamanho da figura (largura e altura).
fig = plt.figure(figsize=(12, 6))

# Cria o objeto que irá armazenar os gráficos ou mapas.
grid = plt.GridSpec(nrows=2, ncols=3)

# Define a disposição das figuras.
ax1 = plt.subplot(grid[0, 0:2])
ax2 = plt.subplot(grid[1, 0:2])
ax3 = plt.subplot(grid[:, 2])
````

# Algumas possibilidades

## Possibilidade 1:

````python
# Define o tamanho da figura (largura e altura).
fig = plt.figure(figsize=(12, 6))

# Cria o objeto que irá armazenar os gráficos ou mapas.
grid = plt.GridSpec(nrows=2, ncols=2)

# Define a disposição das figuras.
ax1 = plt.subplot(grid[0, 0])
ax2 = plt.subplot(grid[0, 1])
ax3 = plt.subplot(grid[1, :])
````

## Possibilidade 2:

````python
# Define o tamanho da figura (largura e altura).
fig = plt.figure(figsize=(12, 6))

# Cria o objeto que irá armazenar os gráficos ou mapas.
# wspace = distância entre as figuras na direção x.
# hspace = distância entre as figuras na direção y.
grid = plt.GridSpec(nrows=4, ncols=4, wspace=0.25, hspace=0.5)

# Define a disposição das figuras.
ax1 = plt.subplot(grid[0, 0])
ax2 = plt.subplot(grid[0, 1:3])
ax3 = plt.subplot(grid[0, 3])
ax4 = plt.subplot(grid[1:4, 0])
ax5 = plt.subplot(grid[1:4, 1:3])
ax6 = plt.subplot(grid[1, 3])
ax7 = plt.subplot(grid[2:4, 3])
````

# Exemplo com dado real

### Importação de biblioteca

````python
import xarray as xr # Biblioteca para ler e manipular arquivo NetCDF.
````

### Abertura do arquivo

````python
dado_nc = xr.open_dataset("../../dados/cru.nc") # Abertura do arquivo NetCDF. É uma série temporal com 12 tempos.

dado_nc # Mostra o conteúdo do arquivo NetCDF.
````

### Importação das variáveis

````python
chuva = dado_nc['pre'][:,0,0]
evap = dado_nc['pet'][:,0,0]
tmin = dado_nc['tmn'][:,0,0]
tmed = dado_nc['tmp'][:,0,0]
tmax = dado_nc['tmx'][:,0,0]
pressaov = dado_nc['vap'][:,0,0]
````

### Geração das figuras

Inicialmente, **gerar 6 figuras**.

````python
# Define o tamanho da figura (largura e altura).
fig = plt.figure(figsize=(14, 6))

# Cria o objeto que irá armazenar os gráficos.
# wspace = distância entre as figuras na direção x.
# hspace = distância entre as figuras na direção y.
grid = plt.GridSpec(nrows=3, ncols=4, wspace=0.2, hspace=0.6)

ax1 = plt.subplot(grid[0, 1])
ax2 = plt.subplot(grid[0, 2])
ax3 = plt.subplot(grid[1, 0:4])
ax4 = plt.subplot(grid[2, 0])
ax5 = plt.subplot(grid[2, 1:3])
ax6 = plt.subplot(grid[2, 3])

fig.suptitle("Título principal da figura")

# Gráfico 1:
ax1.bar(range(0, 12), chuva)
ax1.set_title("Chuva")
ax1.set_ylabel("(mm/mês)")
ax1.set_xlabel("(Mês)")

# Gráfico 2:
ax2.plot(tmax)
ax2.set_title("Tmax")
ax2.set_ylabel("(°C)")
ax2.set_xlabel("(Mês)")

# Gráfico 3:
ax3.bar(range(0,12), evap)
ax3.set_title("Evap. Potencial")
ax3.set_ylabel("(mm/dia)")
ax3.set_xlabel("(Mês)")

# Gráfico 4:
ax4.plot(pressaov)
ax4.set_title("Pressao de Vapor")
ax4.set_ylabel("(hPa)")
ax4.set_xlabel("(Mês)")

# Gráfico 5:
ax5.plot(tmin)
ax5.plot(tmed)
ax5.plot(tmax)
ax5.set_title("Temperatura")
ax5.set_ylabel("(°C)")
ax5.set_xlabel("(Mês)")

# Gráfico 6:
ax6.plot(chuva)
ax6.set_title("Chuva")
ax6.set_ylabel("(mm/mês)")
ax6.set_xlabel("(Mês)")

plt.savefig("ex01.png")
````
